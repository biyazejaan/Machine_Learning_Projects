```python
import ast
```

```python
from ast import literal_eval
```

```python
def convert(obj):
    L =[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
films['genres'] = films['genres'].apply(convert)
```

```python
films.head()
```

```python
films['keywords'] = films['keywords'].apply(convert)
```

```python
def convert3(obj):
    L =[]
    counter = 0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
films['cast'] = films['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L =[]
    counter = 0
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
films['crew'].apply(fetch_director)
```

```python
films['overview'].apply(lambda x:x.split())

## overview column aik string tha jisy baqi lists ky sath concatinate karny ka liya list mein convert kia
```

```python
films['genres'] = films['genres'].apply(lambda x:[i.replace(' ','') for i in x])
films['cast'] = films['cast'].apply(lambda x:[i.replace(' ','') for i in x])
films['crew'] = films['crew'].apply(lambda x:[i.replace(' ','') for i in x])
films['keywords'] = films['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
```

```python
films.head()
```

```python
films['tags'] = films['genres'] + films['cast'] + films['crew'] + films['keywords']
```

```python
df = films[['id','title','tags']] 
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x:" ".join(x))
```

```python
df.head()
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
df['tags'] = df['tags'].apply(stem)
```

```python

 from sklearn.feature_extraction.text import CountVectorizer
 cv = CountVectorizer(max_features=5000,stop_words='english')
 
```

```python
vectors = cv.fit_transform(df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse =True, key = lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = df[df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse =True, key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(df.iloc[i[0]].title)
    return 
```

```python
recommend('Spectre')
```
